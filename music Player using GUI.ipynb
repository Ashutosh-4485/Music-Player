```python
from tkinter import *
from tkinter.ttk import *
from tkinter import filedialog
import time
import pygame
from mutagen.mp3 import MP3
import tkinter.ttk as ttk
root=Tk()
root.title("Music Player")
root.iconbitmap(r'C:\Users\om\forward_button.jpg')
root.geometry('600x600')

#intialize pygame for player
pygame.mixer.init()

#will tell current time of the song
def play_time():
    
    #check for double the timing
    if stopped:
        return
    
    current_time=pygame.mixer.music.get_pos()/1000
    # temporary labl of slider for song
    #slider_label.config(text=f'Slider Pos:{int(my_slider.get())} of Song pos:{int(current_time)}')
    slider_label.config(text=f'Slider Pos:{int(my_slider.get())}')
    
    #IT WILL SHOW IN HOUR,MIN,SEC FORMAT
    
    #status_bar.config(text=int(current_time)) #IT WILL SHOW IN SEC
    convert_cur_time=time.strftime('%M:%S',time.gmtime(current_time))
    status_bar.config(text=convert_cur_time)
    
    #calculating the length of the song
    cur_song=song_box.curselection()
    song=song_box.get(ACTIVE)
    
    #loading the song in mutagen
    song_muta=MP3(song)
    global song_len
    song_len=song_muta.info.length
    convert_song_len=time.strftime('%M:%S',time.gmtime(song_len))
    current_time+=1
    
    if(int(my_slider.get()))==int(song_len):
        status_bar.config(text=f'Time Passed : {convert_song_len}of {convert_song_len}')
        
    elif paused:
        pass
    
    elif(int(my_slider.get()))==int(current_time):
        #slider has moved
        #update slider
        slider_post=int(song_len)
        my_slider.config(to=slider_post,value=int(current_time))
    else:
        #doesn't moved
        #update slider
        slider_post=int(song_len)
        my_slider.config(to=slider_post,value=int(my_slider.get()))
        convert_cur_time=time.strftime('%M:%S',time.gmtime(my_slider.get()))
        status_bar.config(text=f'Time Passed : {convert_cur_time}of {convert_song_len}')
        
        #update it with one sec
        next_time=int(my_slider.get())+1
        my_slider.config(value=next_time)
    
    
    
    #status_bar.config(text=f'Time Passed : {convert_cur_time}of {convert_song_len}') 
    
    #update slider position with the song
    #my_slider.config(value=int(current_time))
                      
    
    #my_slider.config(to=slider_post,value=current_time)
    
   
                      
    #update label using  after  function
    status_bar.after(1000,play_time)
    
    
    
    

       

#slider for the song
def slide(x):
    #slider_label.config(text=f'{int(my_slider.get())} of {int(song_len)}')
    song=song_box.get(ACTIVE)
    #song=f'E:/Anime/audios/{song}.mp3'
    pygame.mixer.music.load(song)
    pygame.mixer.music.play(loops=0,start=int(my_slider.get()))
    
#add song option
def add_song():
    song=filedialog.askopenfilename(initialdir="",title="Choose a song",filetypes=(("mp3 files","*.mp3"),("mp4 files","*.mp4")))
    #Leave initialdir blank it would be better
    #song=song.replace("E:/Anime/audios/"," ")
    #song=song.replace(".mp3"," ")
    song_box.insert(END,song)

#add many song option
def add_multiple_song():
    songs=filedialog.askopenfilenames(initialdir="",title="Choose multiptiple song",filetypes=(("mp3 files","*.mp3"),("mp4 files","*.mp4")))
    for song in songs:
        song_box.insert(END,song)
        
#delete song
def Delete_song():
    stop()
    song_box.delete(ANCHOR)
    pygame.mixer.music.stop()

def Delete_multiple_song():
    song_box.delete(0,END)
    pygame.mixer.music.stop()
    stop()

#Play the song
def play():
    
    global stopped
    stopped=False
    song=song_box.get(ACTIVE)
    #song=f'E:/Anime/audios/{song}.mp3'
    pygame.mixer.music.load(song)
    pygame.mixer.music.play(loops=0)
    #calling for the timing of the song
    play_time()
    #update slider
    #slider_post=int(song_len)
    #my_slider.config(to=slider_post,value=0)

global stopped
stopped=False
    
def stop():
    #reset the slider and all after pressing stop button
    status_bar.config(text='')
    my_slider.config(value=0)
    
    pygame.mixer.music.stop()
    song_box.selection_clear(ACTIVE)
    status_bar.config(text='')
    
    global stopped
    stopped=True

def next_song():
    status_bar.config(text='')
    my_slider.config(value=0)
    #selecting the num of the song
    next_one=song_box.curselection()
    #adding 1 into it
    next_one=next_one[0]+1
    song=song_box.get(next_one)
    pygame.mixer.music.load(song)
    pygame.mixer.music.play(loops=0)
    #first clear the current selection
    song_box.selection_clear(0,END)
    #Activate the next song
    song_box.activate(next_one)
    #moving the bar along with it
    song_box.selection_set(next_one)
    
def previous_song():
    status_bar.config(text='')
    my_slider.config(value=0)
     #selecting the num of the song
    last_one=song_box.curselection()
    #subtracting 1 from it
    last_one=last_one[0]-1
    song=song_box.get(last_one)
    pygame.mixer.music.load(song)
    pygame.mixer.music.play(loops=0)
    #first clear the current selection
    song_box.selection_clear(0,END)
    #Activate the next song
    song_box.activate(last_one)
    #moving the bar along with it
    song_box.selection_set(last_one)
    
#making a global variable to check paused unpaused
global paused
paused=False

def pause(is_paused):
    
    global paused
    paused=is_paused
    if paused==True:
        #unpause the song
        pygame.mixer.music.unpause()
        paused=False
    else:
        # pause the song
        pygame.mixer.music.pause()
        paused=True
    
    
#create playlist box
song_box=Listbox(root,bg='grey',fg='white',width=200,height=30,selectforeground="black")#,selectbackground="green"
song_box.pack(padx=20,pady=20,)


#crate frame for buttons
btn_frame=Frame(root)
btn_frame.pack()

#create  button
back_btn=Button(btn_frame,text="Back",command=previous_song)
forward_btn=Button(btn_frame,text="Next",command=next_song)
play_btn=Button(btn_frame,text="Play",command=play)
stop_btn=Button(btn_frame,text="Stop",command=stop)
pause_btn=Button(btn_frame,text="Pause",command=lambda:pause(paused))

back_btn.grid(row=0,column=0,padx=10)
forward_btn.grid(row=0,column=4,padx=10)
play_btn.grid(row=0,column=2,padx=10)
stop_btn.grid(row=0,column=3,padx=10)
pause_btn.grid(row=0,column=1,padx=10)

#create menu
my_menu=Menu(root)
root.config(menu=my_menu)

#Add Add_song to menu
add_song_opt=Menu(my_menu)
my_menu.add_cascade(label="Add Song",menu=add_song_opt,)
add_song_opt.add_command(label="Add Song",command=add_song)

#Add Add multiple songs to menu
add_song_opt.add_command(label="Add Multiple Songs",command=add_multiple_song)

#create  delete menu
#del_menu=Menu(root)
#root.config(menu=del_menu)

#delete_song 
del_song_opt=Menu(my_menu)
my_menu.add_cascade(label="Delete Song",menu=del_song_opt,)
del_song_opt.add_command(label="Delete Song",command=Delete_song)
#delete multiple songs 
del_song_opt.add_command(label="Delete Multiple Songs",command=Delete_multiple_song)

#create timer for the current song
status_bar=Label(root,text="",relief=GROOVE,anchor=E)
status_bar.pack(fill=X,side=BOTTOM,ipady=2)

#creating the slider
my_slider=ttk.Scale(root,from_=0,to=100,orient=HORIZONTAL,command=slide,length=1200)
my_slider.pack(pady=30)
slider_label=Label(root,text='0')
slider_label.pack(pady=5)



mainloop()
```
